# **Introduction**

Identifying spam emais is an important aspect of both personal and organizational cybersecurity. Identity theft, financial losses, data breaches, significant risks to personal and co-operate security can be caused by malware or phishing scams. They are carried out by spam emails. Organizations use different types of information systems such as CRM systems, ERP systems integrated with email systems. Incidents involving phishing emails can cause a severe threat to the internal functions business. By introducing effective mechanisms for spam email detection, a smoother operation of email systems and preserves bandwidth can be achieved. Integrating updating the organizational cyber security levels constantly will help to keep up with the evolving divercity of spam tactics.  


# **Existing work**

Various machine learning (ML) techniques can be used to achieve the highest accuracy and efficiency of spam detection systems. This section is an evaluation which summarizes the possible ML approaches that can be used for spam email detection by their effectiveness based on existing research.

**Naive Bayes Classifier**

It is one of the earliest techniques used for spam detection. It operates on Bayes' theorem. It is a simple algorithm which  performs well with textual data and is efficient in classifying emails based on word frequencies and other features. According to the study that was conducted it was identified that it can achieve good accuracy with relatively low computational cost (Metsis et al., 2006) .

**Support Vector Machines (SVM)**

It has the potential to handle high-dimensional data. It finds the optimal hyperplane that separates spam from non-spam emails. As indicated by studies specifically used with kernel functions that map input data into higher-dimensional spaces, it reaches high accuracy values (Drucker et al., 1999) .

**Decision Trees and Random Forests**

 Decision Trees learn decision rules. Data features are used for this process.That information will be used to classify spam emails. It makes Decision trees and random forests robust and interpretable. Random Forests is an ensemble method of Decision Trees. It aggregates multiple trees' and produces an output to improve accuracy. Overfitting will be reduced as well. It is capable in handling complex, non-linear relationships in data (Carreras & Márquez, 2001).

**Hybrid Approaches**

Hybrid approaches are created by combining multiple ML techniques. This approch helps to leverage strength of each individual ML technique(base model) that are used to build the ensemble model. This nvolves ensemble techniques that combine predictions from various classifiers or integrate different feature extraction methods to enhance detection accuracy. For instance, combining Naive Bayes with SVMs or using a mixture of neural network architectures has shown improved performance in several studies (Gómez Hidalgo et al., 2006) .

the evolution of ML techniques has improved spam email detection significantly. Each method offers unique advantages.ML techniqies such as Naive Bayes, SVMs, Decision Trees, ANNs, hybrid models, and advanced NLP techniques all contribute to robust spam detection.

#***Support Vector Machines (SVM)***
### **Strengths**:
When the number of features increase, the volume of the space increases exponentially. it makes the data points sparse and the models more complex. This may cause overfitting and computational inefficiency
SVM is designed to handle this efectively.Instead of the data points, it focusses on the margins of separation between classes which makes them less prone to overfitting, even when the dimensionality increases.

### **Weaknesses**:
1. Computationally intensive for large datasets.
2. Parameter tuning and selection of kernel functions should be done carefully

### **Input Data**:
Emails transformed into high-dimensional feature spaces using techniques like TF-IDF (Term Frequency-Inverse Document Frequency).

### **Expected Output**:
A decision boundary wich seperates spam and non-spam emails.

### **Example Application:**
Using SVM to classify emails based on a comprehensive feature set derived from both textual content and metadata (e.g., sender's address, subject line). SVM can effectively separate spam from non-spam emails by finding an optimal hyperplane.

## ***Multinomial Naive Bayes***
It is a probabilistic machine learning algorithm. more suited for classification with features, such as word counts in text classification (descrete features).
Assumption is that features are following a multinomial distribution. Tasks where data is represented as frequency counts are made more effective by this.

### **Strengths**:
1. Training and prediction process is fast and scalable, (suitable for large datasets and large vocabularies).
2. Simplicity: Implementation is easy to understand
3. Performance: Works well for text classification problems, such as spam detection and document classification.

### **Weaknesses**:
1. In real world, the features of the datasets can be conditionally dependant (given the class label). The algorithm makes an assumption which is opposite to this fact. Known as Independant assumption. It assumes that features and labels are not conditionally dependant.
2. Data Representation: If the assumption of multinomial distribution for features does not hold, the performence of alogorithm declines.

### **Input Data**:
Discrete data, often represented as counts or frequencies.
Example: In text classification, the input data would be a document-term matrix where rows represent documents and columns represent word counts.

### **Expected Output**:
A Class label (Spam email or non-spam email)
Example: For spam detection, the output would be a binary label indicating whether an email is spam (1) or not spam (0).

### **Example application**:
Used to classify emails as spam or not spam based on the frequency of words within the emails. By training on a labeled dataset of emails, the algorithm learns the probability of certain words appearing in spam versus non-spam emails and uses this knowledge to classify new emails.

# ***Ensemble Models:***
### **Strengths:**
1. Accuracy can be improved by combining multiple ML models. These  methods often achieve higher predictive accuracy compared to single models. Its achieved through averaging (in bagging) or by focusing on errors made by previous models (in boosting).
2. Robustness: compared to individual models, (Ex: methods such as  random forests and gradient boosting are more robust to noisy data and outliers.
 Flexibility: combinations of different algorithms can be used to handle a wide variety of data types and problem complexities.
3. Stability: Provides more stable and reliable outputs by reducing the variance of predictions.

### **Weaknesses:**
1. Ensemble models can be complex, difficult to interpret and understand.
2. Training and prediction times become generally longer due to different combinations of multiple models. This can be a concerning problem when large datasets are used or in real-time applications.
3. Keeping track of multiple models and their parameters can make difficulties in usage.
4. Updating and maintaining an ensemble require significant effort.
5. Ensembles can require more memory and processing power.

### **Input Data:**
Ensemble models work with numerical, categorical, and textual data.
The preprocessing steps can depend on the individual algorithms used in the ensemble. Generally, the input data should be cleaned, scaled, and appropriately transformed to fulfil the requirements of the models.

### **Expected Output:**
The aggregated prediction of the individual models will be the output of the ensemble model. This might be either,
1. A class label for classification tasks. Determined by either hard voting (majority voting) or soft voting (Probability averaging) or,
2. Mean or weighted average of the predictions  for the regression tasks.

### **Example Application:**
1. Spam Detection
2. Finance risk assessment and credit scoring (reducing the risk of false positives and negatives.)
3. Provide more reliable predictions IN medical imaging and diagnosis.
4. Enhancing recommendation engines.

## **Data Pre-processing**

**The following steppes were followed during the data pre-processing phase:**
1. Reducing words to their root form (Ex: “running” becomes “Run”). This is achieved by using PorterScanner which is a stemming algorithm from the Natural Language Toolkit (NLTK) library.
2. Filtering out common words from the dataset. This is done by using “stopwords” which is a module of NLTK. It provides a list of common words called stopwords). The list of stopwords will then be converted into a set. It helps speedup the testing process.
3. Handle punctuation with string operations. This is achieved by using the standard python library module “string”
4. Convert all characters in the text (email message text) to lowercase. This ensures the text is uniform, (Ex: "Apple" and "apple" will be treated as the same word.)
5. The punctuations will be replaced with “ ”, converted into a translation table, then words broken down separated based on the white spaces.)
6. After these techniques are applied, the list of words (each word separated by white space) will be joined back into single string.
7. The final processed text output will be lowercase, punctuation free. Stemmed and free of stopwords.
8. The data will be handled in a dataframe
9. A new column “transformed-text” will be added to the data frame and every email message text will be mapped to its pre-processed version.
10.	This data frame will be iterated and the preprocessed text was collected into a list “corpus”. This was used for the feature extraction process.

All these tasks was arranged inside the method “**Pre_process(text)**” for the convenience of using where ever needed during the implementation



```python
#reducing the number of words by identifying the word stem
#create a function for this line of code, otherwise it will have to be repeated everyting
stemmer = PorterStemmer()
stopwords_set = set(stopwords.words('english'))
'''
for i in range(len(df)):
  text = df['text'].iloc[i].lower() #turning the words to lower case
'''
def pre_process(text):
    #text = text.replace('\r\n', "")
    text = text.lower() #turning the words to lower case
    text = text.translate(str.maketrans('','', string.punctuation)).split() #removing punctuations and spliting individual waords
    text = [stemmer.stem(word) for word in text if word not in stopwords_set] #stem the word, add it to the stopwords_set if its not already there
    text = ' '.join(text) #join the words with spaces
    #print(text)
    return text
```

```python
df['transformed-text']=df['text'].apply(pre_process)
```

```python
#Data Preprocessing
df['text'].apply(lambda x:nltk.word_tokenize(x)) ## Breaking sentence into number of words
df['text'].apply(lambda x:len(nltk.word_tokenize(x))) ## Counting number of words used in a Mail
```

##**Feature Extraction**

The collection of texts (corpus) will be converted into a matrix of token(words) counts. It will count the occurences of each word. This is done by using the **CountVectorizer** in **sklearn**.


**Fitting and transforming the corpus:**

Machine learning algorithms can not process normal words. In order for it to process, the words have to be converted into a **numerical format.** This is done by “**vectorizer.fit_transform**”. To facilitate further processing the output of this method will be converted into a dense numpy array.



**Splitting the data into test and training sets:**

The extracted features created by “countVectorizer” will be represented by **“x”** parameter. The lables corresponding to each text will be represented by **“y”** parameter. Afterwards, these will be split into four random subsets as following:

**x_train** – Training set features

**x_test** – Test set features

**y_train** – Training set labels

**y_test** – Test set labels


```python
vectorizer = CountVectorizer()
x=vectorizer.fit_transform(corpus).toarray()#to make the email texts readable for algorithms
y=df.label_num
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```
